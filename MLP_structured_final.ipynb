```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import f1_score, roc_auc_score
import optuna
import copy
import random
import os
from collections import Counter
import warnings

# Filter out the specific UserWarning about verbose parameter
warnings.filterwarnings("ignore", message="The verbose parameter is deprecated")

# ‚Äî‚Äî‚Äî Helper to fix randomness ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
def seed_everything(seed: int = 42):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark     = False

seed_everything(42)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("‚úÖ Device:", device)

epochs = 100

# ‚Äî‚Äî‚Äî Load RAW data ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî

y_raw = kick['state_binary'].values                     # shape (n,)
X_raw = kick[[
    'country_displayable_name','prelaunch_activated','staff_pick',
    'parent_category','duration','preparation_days','quarter',
    'creator_backing_count','has_video','goal_usd','gender',
    'blurb_word_count','name_word_count'
]].copy()

# log transforms
X_raw['log_goal_usd']              = np.log1p(X_raw.pop('goal_usd'))
X_raw['log_preparation_days']      = np.log1p(X_raw.pop('preparation_days'))
X_raw['log_creator_backing_count'] = np.log1p(X_raw.pop('creator_backing_count'))

numeric_cols     = [
    'log_goal_usd','log_preparation_days',
    'log_creator_backing_count','duration',
    'blurb_word_count','name_word_count'
]
categorical_cols = ['country_displayable_name','parent_category','gender','quarter']
bool_cols        = ['prelaunch_activated','staff_pick','has_video']

# ‚Äî‚Äî‚Äî STEP 1: INITIAL TRAIN/TEST SPLIT ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
X_train_val_raw, X_test_raw, y_train_val, y_test = train_test_split(
    X_raw, y_raw,
    test_size=0.1,
    stratify=y_raw,
    random_state=42
)

# ‚Äî‚Äî‚Äî Model builder ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
def build_model(input_size: int, trial) -> nn.Module:
    n_layers   = trial.suggest_int("n_layers", 2, 5)
    base_h     = trial.suggest_categorical("hidden_size", [256,384,512])
    activation = trial.suggest_categorical("activation", ["ReLU","LeakyReLU"])
    dropout    = trial.suggest_float("dropout", 0.1, 0.4)

    layers = []
    in_f = input_size
    for i in range(n_layers):
        h = max(base_h // (2**i), 32)
        layers.append(nn.Linear(in_f, h))
        layers.append(getattr(nn, activation)() if activation!="LeakyReLU"
                      else nn.LeakyReLU(0.1))
        layers.append(nn.Dropout(dropout))
        in_f = h
    layers.append(nn.Linear(in_f, 1))
    return nn.Sequential(*layers).to(device)

# ‚Äî‚Äî‚Äî Training loop with early‚Äêstopping & pruning ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
def train_model(model, train_loader, val_loader, optimizer, criterion,
                epochs: int, patience: int, trial=None, fold_offset: int=0,
                clip_value=1.0, delta=1e-4):
    best_loss = float('inf')
    wait = 0
    best_state = copy.deepcopy(model.state_dict())

    for epoch in range(epochs):
        # Training phase
        model.train()
        for xb, yb in train_loader:
            optimizer.zero_grad()
            loss = criterion(model(xb), yb)
            loss.backward()

            # Add gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), clip_value)

            optimizer.step()

        # Validation phase
        model.eval()
        val_loss = 0.0
        all_preds, all_labels = [], []
        with torch.no_grad():
            for xb, yb in val_loader:
                outputs = model(xb)
                batch_loss = criterion(outputs, yb)
                val_loss += batch_loss.item() * len(xb)

                p = torch.sigmoid(outputs).cpu()
                all_preds.append(p)
                all_labels.append(yb.cpu())

        # Calculate average validation loss
        val_loss /= len(val_loader.dataset)

        # For Optuna trial reporting
        if trial:
            preds = torch.cat(all_preds).numpy()
            labs = torch.cat(all_labels).numpy()
            f1 = f1_score(labs, (preds>0.5).astype(int))
            trial.report(f1, step=fold_offset+epoch)
            if trial.should_prune():
                raise optuna.exceptions.TrialPruned()

        # Early stopping check based on validation loss (consecutive)
        if val_loss < best_loss - delta:
            best_loss = val_loss
            best_state = copy.deepcopy(model.state_dict())
            wait = 0
        else:
            wait += 1
            if wait >= patience:
                break

    if best_state:
        model.load_state_dict(best_state)
    return model, best_loss

# ‚Äî‚Äî‚Äî STEP 2: NESTED CV ON TRAIN+VAL ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
outer_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
outer_results = []

for fold_idx, (tr_idx, va_idx) in enumerate(
        outer_cv.split(X_train_val_raw, y_train_val), start=1):
    print(f"\nüîÅ Outer fold {fold_idx}")

    # Split raw
    X_tr_raw = X_train_val_raw.iloc[tr_idx]
    y_tr     = y_train_val[tr_idx]
    X_va_raw = X_train_val_raw.iloc[va_idx]
    y_va     = y_train_val[va_idx]

    # Fit transformer on X_tr_raw only
    pre = ColumnTransformer([
        ("num",   StandardScaler(), numeric_cols),
        ("cat",   OneHotEncoder(sparse_output=False,
                                handle_unknown="ignore"),
                   categorical_cols),
        ("bools","passthrough", bool_cols),
    ])
    pre.fit(X_tr_raw)

    # Transform once for outer train/val
    X_tr = pre.transform(X_tr_raw)
    X_va = pre.transform(X_va_raw)

    # Inner‚ÄêCV objective
    def objective(trial):
        model = build_model(X_tr.shape[1], trial)
        opt_name = trial.suggest_categorical("optimizer", ["Adam","AdamW"])
        lr       = trial.suggest_float("lr",1e-5,1e-3,log=True)
        wd       = trial.suggest_float("weight_decay",1e-6,1e-3,log=True)
        bs       = trial.suggest_categorical("batch_size",[128,256,512])

        pat = 7
        criterion = nn.BCEWithLogitsLoss()

        skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
        scores = []

        for i, (it, iv) in enumerate(skf.split(X_tr, y_tr)):
            Xi_tr, Xi_va = X_tr[it], X_tr[iv]
            yi_tr, yi_va = y_tr[it], y_tr[iv]

            dl_tr = DataLoader(TensorDataset(
                torch.tensor(Xi_tr, dtype=torch.float32).to(device),
                torch.tensor(yi_tr, dtype=torch.float32).unsqueeze(1).to(device)
            ), batch_size=bs, shuffle=True)
            dl_va = DataLoader(TensorDataset(
                torch.tensor(Xi_va, dtype=torch.float32).to(device),
                torch.tensor(yi_va, dtype=torch.float32).unsqueeze(1).to(device)
            ), batch_size=bs)

            m_copy = copy.deepcopy(model)
            optimizer = getattr(optim, opt_name)(
                m_copy.parameters(), lr=lr, weight_decay=wd
            )

            trained, _ = train_model(
                m_copy, dl_tr, dl_va,
                optimizer, criterion,
                epochs=epochs, patience=pat,
                trial=trial, fold_offset=i*epochs,
                clip_value=1.0, delta=1e-4
            )

            trained.eval()
            with torch.no_grad():
                p = torch.sigmoid(trained(
                    torch.tensor(Xi_va, dtype=torch.float32).to(device)
                )).cpu().numpy()
            scores.append(f1_score(yi_va, (p>0.5).astype(int)))

        return np.mean(scores)

    study = optuna.create_study(
        direction="maximize",
        pruner=optuna.pruners.MedianPruner(
            n_startup_trials=5, n_warmup_steps=5
        )
    )
    study.optimize(objective, n_trials=30)

    best = study.best_trial.params
    print(" ‚Üí Best params:", best)

    # Retrain on full outer‚Äêtrain
    final_model = build_model(X_tr.shape[1], optuna.trial.FixedTrial(best))
    optimizer   = getattr(optim, best['optimizer'])(
        final_model.parameters(),
        lr=best['lr'], weight_decay=best['weight_decay']
    )
    criterion   = nn.BCEWithLogitsLoss()

    dl_tr = DataLoader(TensorDataset(
        torch.tensor(X_tr, dtype=torch.float32).to(device),
        torch.tensor(y_tr, dtype=torch.float32).unsqueeze(1).to(device)
    ), batch_size=best['batch_size'], shuffle=True)
    dl_va = DataLoader(TensorDataset(
        torch.tensor(X_va, dtype=torch.float32).to(device),
        torch.tensor(y_va, dtype=torch.float32).unsqueeze(1).to(device)
    ), batch_size=best['batch_size'])

    final_model, _ = train_model(
        final_model, dl_tr, dl_va,
        optimizer, criterion,
        epochs=epochs, patience=7,
        clip_value=1.0, delta=1e-4
    )

    # Evaluate on outer‚Äêval
    final_model.eval()
    with torch.no_grad():
        p = torch.sigmoid(final_model(
            torch.tensor(X_va, dtype=torch.float32).to(device)
        )).cpu().numpy()
    f1 = f1_score(y_va, (p>0.5).astype(int))
    auc = roc_auc_score(y_va, p)
    print(f"Fold {fold_idx} ‚Üí F1={f1:.4f}, AUC={auc:.4f}")

    outer_results.append({
        "f1": f1, "auc": auc,
        "params": best,
        "state": copy.deepcopy(final_model.state_dict())
    })

#‚Äî‚Äî‚Äî STEP 3: FINAL TRAIN+TEST EVAL ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî

# 3a) select the best fold based on performance

best_fold_idx = np.argmax([r["f1"] for r in outer_results])
best_fold = outer_results[best_fold_idx]
best_params = best_fold["params"]

print(f"üìå Using parameters from fold {best_fold_idx + 1} with F1={best_fold['f1']:.4f}, AUC={best_fold['auc']:.4f}")
print("üìå Best hyperparameters:", best_params)

# 3b) Re-fit preprocessor on all train+val raw, transform train+val and test
final_pre = ColumnTransformer([
    ("num",   StandardScaler(), numeric_cols),
    ("cat",   OneHotEncoder(sparse_output=False, handle_unknown="ignore", drop=None),
               categorical_cols),
    ("bools","passthrough", bool_cols),
])
final_pre.fit(X_train_val_raw)

X_tv = final_pre.transform(X_train_val_raw)   # numpy array
X_te = final_pre.transform(X_test_raw)

# 3c) Build & train the final model on (X_tv, y_train_val)

# prepare fixed-trial for Optuna style
fixed_trial = optuna.trial.FixedTrial(best_params)
final_model = build_model(X_tv.shape[1], fixed_trial)

optimizer = getattr(optim, best_params["optimizer"])(
    final_model.parameters(),
    lr=best_params["lr"],
    weight_decay=best_params["weight_decay"]
)
criterion = nn.BCEWithLogitsLoss()

# convert to tensors and make DataLoader
X_tv_tensor = torch.tensor(X_tv, dtype=torch.float32).to(device)
y_tv_tensor = torch.tensor(y_train_val, dtype=torch.float32).unsqueeze(1).to(device)

# small 10% val split from train+val for early stopping
idx_train, idx_val = train_test_split(
    np.arange(len(X_tv_tensor)),
    test_size=0.1,
    stratify=y_train_val,
    random_state=42
)

train_loader = DataLoader(
    TensorDataset(X_tv_tensor[idx_train], y_tv_tensor[idx_train]),
    batch_size=best_params["batch_size"], shuffle=True
)
val_loader = DataLoader(
    TensorDataset(X_tv_tensor[idx_val], y_tv_tensor[idx_val]),
    batch_size=best_params["batch_size"]
)

# Modified train_model function: tracking training/validation loss and F1 scores
# Modified train_model_with_metrics function to use val_loss for early stopping
def train_model_with_metrics(model, train_loader, val_loader, optimizer, criterion,
                           epochs: int, patience: int, clip_value=1.0, delta=1e-4):
    best_loss = float('inf')  # Initialize with infinity for loss minimization
    wait = 0
    best_state = copy.deepcopy(model.state_dict())
    best_epoch = 0

    # Lists to track metrics
    train_losses = []
    val_losses = []
    train_f1s = []
    val_f1s = []

    for epoch in range(epochs):
        # Training phase
        model.train()
        train_loss = 0.0
        train_preds = []
        train_labels = []

        for xb, yb in train_loader:
            optimizer.zero_grad()
            outputs = model(xb)
            loss = criterion(outputs, yb)
            loss.backward()


            torch.nn.utils.clip_grad_norm_(model.parameters(), clip_value)

            optimizer.step()

            train_loss += loss.item() * len(xb)
            p = torch.sigmoid(outputs).cpu().detach()
            train_preds.append(p)
            train_labels.append(yb.cpu().detach())

        # Calculate training loss and F1
        train_loss /= len(train_loader.dataset)
        train_preds = torch.cat(train_preds).numpy()
        train_labels = torch.cat(train_labels).numpy()
        train_f1 = f1_score(train_labels, (train_preds > 0.5).astype(int))

        # Validation phase
        model.eval()
        val_loss = 0.0
        val_preds = []
        val_labels = []

        with torch.no_grad():
            for xb, yb in val_loader:
                outputs = model(xb)
                loss = criterion(outputs, yb)
                val_loss += loss.item() * len(xb)

                p = torch.sigmoid(outputs).cpu()
                val_preds.append(p)
                val_labels.append(yb.cpu())

        # Calculate validation loss and F1
        val_loss /= len(val_loader.dataset)
        val_preds = torch.cat(val_preds).numpy()
        val_labels = torch.cat(val_labels).numpy()
        val_f1 = f1_score(val_labels, (val_preds > 0.5).astype(int))

        # Store metrics
        train_losses.append(train_loss)
        val_losses.append(val_loss)
        train_f1s.append(train_f1)
        val_f1s.append(val_f1)

        # Print progress
        print(f"Epoch {epoch+1}/{epochs} - Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, "
              f"Train F1: {train_f1:.4f}, Val F1: {val_f1:.4f}")

        # Early stopping check (based on validation loss)
        if val_loss < best_loss - delta:
            best_loss = val_loss
            best_state = copy.deepcopy(model.state_dict())
            best_epoch = epoch
            wait = 0
        else:
            wait += 1
            if wait >= patience:
                print(f"Early stopping triggered! No improvement for {patience} consecutive epochs.")
                print(f"Best epoch was {best_epoch+1} with validation loss: {best_loss:.4f}")
                break

    # Restore best model after training
    if best_state:
        model.load_state_dict(best_state)

    # Visualization code remains the same...
    import matplotlib.pyplot as plt

    plt.figure(figsize=(15, 10))

    # Loss curves
    plt.subplot(2, 1, 1)
    plt.plot(train_losses, label='Training Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.axvline(x=best_epoch, color='r', linestyle='--',
                label=f'Best Epoch ({best_epoch+1})')
    plt.title('Loss Curves')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    # F1 score curves
    plt.subplot(2, 1, 2)
    plt.plot(train_f1s, label='Training F1')
    plt.plot(val_f1s, label='Validation F1')
    plt.axvline(x=best_epoch, color='r', linestyle='--',
                label=f'Best Epoch ({best_epoch+1})')
    plt.title('F1 Score Curves')
    plt.xlabel('Epoch')
    plt.ylabel('F1 Score')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # Return model, best loss, and metrics
    return model, best_loss, {
        'best_epoch': best_epoch,
        'train_losses': train_losses,
        'val_losses': val_losses,
        'train_f1s': train_f1s,
        'val_f1s': val_f1s
    }

# Train the final model with metrics tracking
print("\nüîÑ Training final model with metrics tracking...")
final_model, best_val_loss, metrics = train_model_with_metrics(
    final_model, train_loader, val_loader,
    optimizer, criterion,
    epochs=epochs, patience=7,
    clip_value=1.0, delta=1e-4
)

print(f"\nüìä Final model training completed!")
print(f"üìâ Best validation loss: {best_val_loss:.4f} at epoch {metrics['best_epoch']+1}")

# Calculation of metric
final_model.eval()
with torch.no_grad():
    # train set performance
    train_outputs = torch.sigmoid(final_model(X_tv_tensor[idx_train])).cpu().numpy()
    train_f1_final = f1_score(y_train_val[idx_train], (train_outputs > 0.5).astype(int))
    train_auc_final = roc_auc_score(y_train_val[idx_train], train_outputs)

    # test set performance
    val_outputs = torch.sigmoid(final_model(X_tv_tensor[idx_val])).cpu().numpy()
    val_f1_final = f1_score(y_train_val[idx_val], (val_outputs > 0.5).astype(int))
    val_auc_final = roc_auc_score(y_train_val[idx_val], val_outputs)

print(f"\nüìä Final Model Performance:")
print(f"Training   ‚Üí F1={train_f1_final:.4f}, AUC={train_auc_final:.4f}")
print(f"Validation ‚Üí F1={val_f1_final:.4f}, AUC={val_auc_final:.4f}")

# Learning curve analysis
print("\nüìâ Learning Curve Analysis:")
total_epochs = len(metrics['train_losses'])
print(f"Total epochs trained: {total_epochs}")
print(f"Early stopping occurred at epoch {metrics['best_epoch']+1} (patience=5)")

# Metrics at early stopping point
es_epoch = metrics['best_epoch']
print(f"\nMetrics at the Early Stopping point (Epoch {es_epoch+1}):")
print(f"Training Loss: {metrics['train_losses'][es_epoch]:.4f}")
print(f"Validation Loss: {metrics['val_losses'][es_epoch]:.4f}")
print(f"Training F1: {metrics['train_f1s'][es_epoch]:.4f}")
print(f"Validation F1: {metrics['val_f1s'][es_epoch]:.4f}")
```

```python
# After completing the final model training on train+val set, now evaluate on the test set
print("\nüß™ Evaluating the final model on the test set...")

# Convert test set to tensors
X_test_tensor = torch.tensor(X_te, dtype=torch.float32).to(device)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32).unsqueeze(1).to(device)

# Create test dataloader
test_loader = DataLoader(
    TensorDataset(X_test_tensor, y_test_tensor),
    batch_size=best_params["batch_size"]
)

# Evaluate on test set
final_model.eval()
test_loss = 0.0
test_preds = []
test_labels = []

with torch.no_grad():
    for xb, yb in test_loader:
        outputs = final_model(xb)
        loss = criterion(outputs, yb)
        test_loss += loss.item() * len(xb)

        p = torch.sigmoid(outputs).cpu()
        test_preds.append(p)
        test_labels.append(yb.cpu())

# Calculate test metrics
test_loss /= len(test_loader.dataset)
test_preds = torch.cat(test_preds).numpy()
test_labels = torch.cat(test_labels).numpy()

# Binary predictions (using 0.5 threshold)
test_pred_binary = (test_preds > 0.5).astype(int)

# Calculate metrics
test_f1 = f1_score(test_labels, test_pred_binary)
test_auc = roc_auc_score(test_labels, test_preds)

print(f"\nüìä Test Set Performance:")
print(f"Test Loss: {test_loss:.4f}")
print(f"Test F1 Score: {test_f1:.4f}")
print(f"Test AUC: {test_auc:.4f}")

# Confusion Matrix and Classification Report (optional, provides more detailed metrics)
from sklearn.metrics import confusion_matrix, classification_report

# Confusion matrix
conf_matrix = confusion_matrix(test_labels, test_pred_binary)
print("\nConfusion Matrix:")
print(conf_matrix)

# Classification report
class_report = classification_report(test_labels, test_pred_binary)
print("\nClassification Report:")
print(class_report)

# Optional: Save the model
torch.save({
    'model_state_dict': final_model.state_dict(),
    'optimizer_state_dict': optimizer.state_dict(),
    'best_params': best_params,
    'preprocessor': final_pre,
    'test_performance': {
        'f1': test_f1,
        'auc': test_auc,
        'loss': test_loss
    }
}, 'kickstarter_prediction_model.pt')

print("\nüíæ Model saved to 'kickstarter_prediction_model.pt'")
```
